### Importing necessary Modules ###

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as  plt
import seaborn as sns
from sklearn import datasets
```

### Loading the dataset ###

```python
df=datasets.load_diabetes(scaled=False)
x=df.data
y=df.target
columns=df.feature_names
```

```python
data=pd.DataFrame(x,columns=columns)
data
```

### Deetecting & Handling Outliers ###

```python
sns.boxplot(data)
plt.show()
```

```python
drp = data.copy()
for i in range(data.shape[1]):
    feature_data = data.iloc[:,i]
    low=np.percentile(feature_data,10)
    up=np.percentile(feature_data,90)
    drp.iloc[feature_data.index[feature_data < low],i]=low
    drp.iloc[feature_data.index[feature_data > up],i]=up
```

### Feature Selection Using Filter Method ###

```python
from sklearn.feature_selection import mutual_info_classif,chi2,SelectKBest,mutual_info_regression
from sklearn.metrics import accuracy_score
from skfeature.function.similarity_based import fisher_score
```

```python
inf=mutual_info_regression(data,y,random_state=42)
info=pd.DataFrame(inf,columns)
info.plot(kind='barh',color='pink')
plt.show()
```

```python
xs=np.array(data)
inf=fisher_score.fisher_score(xs,y)
info=pd.DataFrame(inf,columns)
info.plot(kind='barh',color='pink')
plt.show()
```

```python
dx=SelectKBest(chi2,k=2)
dx.fit(data,y)
indi=dx.get_support(indices=True)
data.columns[indi]
```

### Divide the data in training and testing set ###

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data,y,random_state=42,test_size=0.3)
```

### Implementing various Regression model ###

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
l=LinearRegression()
l.fit(xtrain,ytrain)
ypred=l.predict(xtest)
r2_score(ytest,ypred)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k=KNeighborsRegressor(n_neighbors=12)
k.fit(xtrain,ytrain)
ypred=k.predict(xtest)
r2_score(ytest,ypred)
```

```python
from sklearn.svm import SVR,SVC
```

```python
sv = SVC()
sv.fit(xtrain,ytrain)
ypredd=sv.predict(xtest)
r2_score(ytest,ypred)
```

```python
s=SVR()
s.fit(xtrain,ytrain)
ypred=s.predict(xtest)
r2_score(ytest,ypred)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
d=DecisionTreeRegressor()
d.fit(xtrain,ytrain)
ypred=d.predict(xtest)
r2_score(ytest,ypred)
```

```python
plot_tree(d)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
r=RandomForestRegressor()
r.fit(xtrain,ytrain)
ypred=r.predict(xtest)
r2_score(ytest,ypred)
```

### Cross validation using Kfold and LOO  ###

```python
from sklearn.model_selection import KFold,LeaveOneOut,cross_val_score,GridSearchCV
```

```python
kf=KFold(n_splits=3)
kl=cross_val_score(l,data,y,cv=kf)
kk=cross_val_score(k,data,y,cv=kf)
ks=cross_val_score(s,data,y,cv=kf)
kd=cross_val_score(d,data,y,cv=kf)
kr=cross_val_score(r,data,y,cv=kf)
print(kl.mean(),kk.mean(),ks.mean(),kd.mean(),kr.mean())
```

```python
wcss=[]
for i in range(1,11):
    kma=KMeans(n_clusters=i)
    kma.fit(data)
    wcss.append(kma.inertia_)
plt.plot(range(1,11),wcss)
plt.show()
```
